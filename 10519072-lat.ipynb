```python
import matplotlib.pyplot as plt 
import numpy as np 
import pandas as pd 

x = df['sepal_length'].loc[0:10]
y = df['sepal_width'].loc[0:10]

df = pd.DataFrame({ 'sepal_length': x, 'sepal_width': y, }) 
df.sort_values(by='sepal_width', inplace=True) 
x_coords = np.arange(len(df)) 
plt.bar(x_coords, df['sepal_width'], tick_label=df['sepal_length']) 
plt.xticks(rotation=90) 
plt.ylabel('Sepal Width') 
plt.title('Sepal Length') 
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np 
import pandas as pd 

x = df['sepal_length'].loc[0:10]
y = df['sepal_width'].loc[0:10]

df = pd.DataFrame({ 'sepal_length': x, 'sepal_width': y, }) 
df.sort_values(by='sepal_width', inplace=True) 
x_coords = np.arange(len(df)) 
colors = ['#0000FF' for _ in range(len(df))] 
colors[-2] = '#FF0000' 
plt.bar(x_coords, df['sepal_width'], tick_label=df['sepal_length'], color=colors) 
plt.xticks(rotation=90) 
plt.ylabel('Sepal Width') 
plt.title('sepal Length') 
plt.show()
```

```python
import matplotlib.pyplot as plt 

x = df['sepal_length'].loc[0:10]
y = df['sepal_width'].loc[0:10]

plt.plot( x, y, ) 
plt.title('Data Iris') 
plt.ylabel('Sepal Length') 
plt.xlabel('Sepal Width') 
plt.show()
```
